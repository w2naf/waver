```python
# This part would not have to be executed if you were running DaViT-py
import sys
sys.path.append('/davitpy')
# But all that comes after this point would probably be needed
from pydarn import *
from mpl_toolkits.basemap import Basemap
from datetime import datetime
from utils.calcSun import *
date = datetime(2012, 6, 17, 21, 15)
jd = getJD(date)
lats, lons, zen, term = calcTerminator( jd, [30.,90.], [-180.,180.] )
```

```python
# setup map projection and decorations
figure(figsize=(15,8))
subplot(131)
map = Basemap(projection='npstere',boundinglat=30,lon_0=270,resolution='l')
map.drawcoastlines()
map.drawmapboundary()
# draw parallels and meridians.
map.drawparallels(np.arange(30.,81.,20.))
map.drawmeridians(np.arange(-180.,181.,20.))
# overplot terminator
map.nightshade(date)
# plot calculated zenith data
subplot(132)
map = Basemap(projection='npstere',boundinglat=30,lon_0=270,resolution='l')
map.drawcoastlines()
map.drawmapboundary()
map.drawparallels(np.arange(30.,81.,20.))
map.drawmeridians(np.arange(-180.,181.,20.))
ny = zen.shape[0]; nx = zen.shape[1]
zendat, x, y = map.transform_scalar(zen,lons,lats,nx,ny, returnxy=True)
map.contourf(x, y, zendat, [90,180])
# plot calculated zenith data
subplot(133)
map = Basemap(projection='npstere',boundinglat=30,lon_0=270,resolution='l')
map.drawcoastlines()
map.drawmapboundary()
map.drawparallels(np.arange(30.,81.,20.))
map.drawmeridians(np.arange(-180.,181.,20.))
ny = zen.shape[0]; nx = zen.shape[1]
zendat, x, y = map.transform_scalar(zen,lons,lats,nx,ny, returnxy=True)
map.imshow(zendat, pylab.cm.afmhot_r, alpha=.9)
x, y = map(term[:,1], term[:,0])
map.plot(x,y)
```
